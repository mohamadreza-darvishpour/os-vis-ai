### Displaying Images

```python
import cv2
import os
import matplotlib.pyplot as plt
# cv2.imread will read the image and convert it into an ndarray.
base_path = '/content/drive/MyDrive/Image Processing and Acquisition using Python/Chapter03'
# Opening the image.
a = cv2.imread(os.path.join(base_path, 'images/Gray.jpg'))
img = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
# We import matplotlib.pyplot to display an image in grayscale.
# If gray is not supplied the image will be displayed in color.
plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plt.imshow(img, 'gray')
x = img.ravel()
plt.subplot(1,2,2)
plt.hist(x,256,[0,255])
plt.show()
```

```python
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np

# cv2.imread will read the image and convert it into an ndarray.
base_path = '/content/drive/MyDrive/Image Processing and Acquisition using Python/Chapter03'
# Opening the image.
a = cv2.imread(os.path.join(base_path, 'images/Gray.jpg'))
img = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
# create the histogram
vals = img.flatten()
# plot histogram with 255 bins
b, bins, patches = plt.hist(vals, bins=255, range=(0,255))
plt.xlim([0,255])
plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt
# cv2.imread will read the image and convert it into an ndarray.
base_path = '/content/drive/MyDrive/Image Processing and Acquisition using Python/Chapter03'
a = Image.open(os.path.join(base_path, 'images/RGB.jpg'))
img = a.convert('L')
# We import matplotlib.pyplot to display an image in grayscale.
# If gray is not supplied the image will be displayed in color.
plt.figure(figsize=(15,8))
plt.subplot(1,2,1)
plt.title('RGB Image')
plt.imshow(a)

plt.subplot(1,2,2)
plt.title('GrayScale Image')
plt.imshow(img, 'gray')
```
